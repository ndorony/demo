```python
from IPython.display import Markdown, display

def print(a):
    display(Markdown(f'<div dir="rtl">{a}</div>'))

def print_trophy(a):
    display(Markdown(f'<pre dir="ltr" style="font-family:monospace">{a}</pre>'))

right = 0
total = 0

def ask(question, a, b, c, real_answer):
    global right, total
    total += 1
    print(question)
    print(f"1. {a}")
    print(f"2. {b}")
    print(f"3. {c}")

    print("בחר תשובה (1-3): ")
    answer = input()

    if answer.strip() and  answer.strip()[0] == str(real_answer):
        right += 1
        print("נכון!")
    else:
        print("לא נכון!")

    print("<br>")


print("ברוכים הבאים למבחן השאלות של יום הולדת של דוד ושקד")
print("<br>")

print("שאלה 1")
ask("בן כמה שקד?", '10', '9', '11', '1')

print("שאלה 2")
ask("איפה שקד גר?", 'ההגנה 3', 'ההגנה 3 על עמודים', 'ההגנה 3 א', '2')

print("שאלה 3")
ask("מה התחביב האהוב על שקד?", "לבנות מחשב", "לפתור קוביות הונגריות", "לעשות אפליקציות", '3')

print("שאלה 4")
ask("באיזה חודש נולד רפאל?",
    "ח' בסיוון תשע\"ו",
    "י' בתמוז תשע\"ו",
    "א' באייר תשע\"ו",
    "1")

print("שאלה 5")
ask("מה הצבע האהוב על רפאל?",
    "כחול",
    "שחור",
    "אדום",
    "2")

print("שאלה 6")
ask("מה רפאל מעדיף?",
    "פיצה",
    "המבורגר",
    "שניהם באותה מידה",
    "2")

print("סוף המבחן! כל הכבוד 🎉")


from IPython.display import HTML, display

def print_trophy(score, label="WINNER", title="!ניצחון"):
    trophy = rf"""
   __                   __
   \ \_   .-======-.  _/ /
    \  \ / * * * * \ /  /
     \ V|{score.center(11)}|V /
      \ |{label.center(11)}| /
       \|           |/
        \  * * * *  /
         '.       .'
           \     /
            |   |
           .'   '.
          /  STAR \
         /_________\
        [===========]
       [=============]
"""
    html = f"""
    <style>
      @keyframes shimmer {{
        0%   {{ background-position: -200% center; }}
        100% {{ background-position:  200% center; }}
      }}
      @keyframes glowpulse {{
        0%, 100% {{ text-shadow: 0 0 8px rgba(255,200,0,.5), 0 0 18px rgba(255,160,0,.35); }}
        50%      {{ text-shadow: 0 0 16px rgba(255,220,80,.9), 0 0 34px rgba(255,180,0,.6); }}
      }}
      .trophy-wrap {{
        display:flex; flex-direction:column; align-items:center;
        padding:24px; border-radius:18px;
        background: radial-gradient(circle at 50% 0%, #2a2410 0%, #14110a 60%, #0a0806 100%);
        box-shadow: inset 0 0 60px rgba(255,180,0,.12), 0 6px 30px rgba(0,0,0,.5);
        width:max-content; margin:0 auto;
      }}
      .trophy-title {{
        font-family: 'Arial Black', sans-serif;
        font-size:30px; font-weight:900; margin-bottom:6px;
        background: linear-gradient(90deg,#a8741a,#ffe17a,#fff6c8,#ffe17a,#a8741a);
        background-size:200% auto;
        -webkit-background-clip:text; background-clip:text;
        -webkit-text-fill-color:transparent;
        animation: shimmer 3s linear infinite;
        letter-spacing:1px;
      }}
      .trophy-art {{
        direction:ltr; unicode-bidi:isolate;
        font-family: 'Courier New', monospace; font-size:16px;
        line-height:1.12; white-space:pre; text-align:left; margin:0;
        background: linear-gradient(180deg,#fff3b0 0%,#ffcf3f 40%,#d99405 100%);
        -webkit-background-clip:text; background-clip:text;
        -webkit-text-fill-color:transparent;
        animation: glowpulse 2.2s ease-in-out infinite;
        filter: drop-shadow(0 0 6px rgba(255,180,0,.4));
      }}
    </style>
    <div class="trophy-wrap">
      <div class="trophy-title">{title}</div>
      <pre class="trophy-art">{trophy}</pre>
    </div>
    """
    display(HTML(html))


score = f"{right}/{total}"
print_trophy(score)


```
